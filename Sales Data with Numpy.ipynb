```python
import pandas as pd
import numpy as np
Sales = pd.read_csv('salesdata.csv')
Sales
```

## Group 1

```python
Sales.head()
```

```python
# Removing the $ sign in the item price
```

```python
Sales['item_price'] = Sales['item_price'].replace('$', '').astype(float)
Sales
```

## How many products cost more than $10.00?

```python
np.count_nonzero(Sales.item_price>10.00)
```

## How many products cost less than $10.00?

```python
np.count_nonzero(Sales.item_price<10.00)
```

# Group 2

## What is the price of each item? Sort the products by the name of the item

```python
# high_products = products.reset_index().groupby(['name', 'brand'])[['name', 'brand', 'count_name']]
product_name_price = Sales.groupby(by = 'item_name')[['item_price']].min().reset_index()
product_name_price
```

## Sort the products by the name of the item

```python
Sales.sort_values(by= 'item_name',ascending=True).reset_index(drop=True)
```

# Group 3

## What was the quantity of the most expensive item ordered?

```python
# df[df['item_price'] == df['item_price'].max()].shape[0]
Sales[Sales['item_price'] == Sales['item_price'].max()].shape[0]
```

## How many times was a Chicken Burrito ordered?

```python
numberofordersforchickenburrito = np.count_nonzero(Sales.item_name == 'Chicken Burrito')
numberofordersforchickenburrito
```

## How many times did someone order more than one Canned Soda?

```python
numberoftimescannedsodamorethan1 = np.count_nonzero((Sales.item_name == 'Chicken Burrito') & (Sales.quantity > 1))
numberoftimescannedsodamorethan1
```
